```python
import json
import numpy as np
```

```python
def preapre_data(json_path: str, test_name: str, features: list[str], TStroop_way = 1):  #TStroop_way = 1 если разделяем mono и trueColor, 2 если нет

    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    X_list = []
    y_list = []
    for rec in data:
        if rec["test_name"] == "TStroop":
            if TStroop_way == 1:
                seq = []
                for row in rec["test_results"]["mono"]: #словарь в словаре в котором список словарей
                    values = []
                    for f in features:
                        values.append(row[f])    #добавляем в values данные по нужным признакам (строка)
                    seq.append(values)       #добавляем в seq строку
                for row in rec["test_results"]["trueColor"]: #словарь в словаре в котором список словарей
                    values = []
                    for f in features:
                        values.append(row[f])    #добавляем в values данные по нужным признакам (строка)
                    seq.append(values)       #добавляем в seq строку
                
                X_list.append(np.array(seq, dtype=float)) #добавляем таблицу по одной попытке теста
                y_list.append(rec["age"])
            if TStroop_way == 2:
                seq = []
                for row in rec["test_results"]["mono"]: #словарь в словаре в котором список словарей
                    values = []
                    for f in features:
                        values.append(row[f])    #добавляем в values данные по нужным признакам (строка)
                for row in rec["test_results"]["trueColor"]: #словарь в словаре в котором список словарей
                    values = []
                    for f in features:
                        values.append(row[f])    #добавляем в values данные по нужным признакам (строка)
                    seq.append(values)       #добавляем в seq строку

                X_list.append(np.array(seq, dtype=float)) #добавляем таблицу по одной попытке теста
                y_list.append(rec["age"])

        else:

            if rec["test_name"] == test_name:  
                seq = []
                for row in rec["test_results"]: #список словарей
                    values = []
                    for f in features:
                        values.append(row[f])    #добавляем в values данные по нужным признакам (строка)
                    seq.append(values)       #добавляем в seq строку
                X_list.append(np.array(seq, dtype=float)) #добавляем таблицу по одной попытке теста
                y_list.append(rec["age"])  

    #  Padding (выравнивание по длине) 
    max_len = max(seq.shape[0] for seq in X_list) #максимальное количество строк в таблице
    num_features = len(features)
    batch_size = len(X_list)

    X = np.zeros((batch_size, max_len, num_features), dtype=float) #итоговый тензор
    for i, seq in enumerate(X_list):
        X[i, :seq.shape[0], :] = seq #заполняем тензор данными, там где их нет, ставим 0

    y = np.array(y_list, dtype=float)
    
    return X, y
```

```python
json_path = "calculator_9_2025-03-19_filter.json"
test_name = ["T1back", "TStroop", "T258", "T274", "T278"]
features_T1back = [
    "Stimul", "Goal", "Color", "Duration",
    "Show", "Hide", "Down", "Up",
    "SMR", "MR", "ERR_1", "ERR_2", "ERR_3"
]
features_TStroop = ["True", "Error", "SMR", "MR", "Time"]
features_T258 = ["Stimul", "H", "H+", "dH+", "H-", "dH-", "t+", "t-", "ERR", "ERR_LIM"]
features_T274 = ["Stimul", "Goal", "Duration", "Interval", "Show", "Hide", "Down", "Up", "SMR", "MR", "ERR_1", "ERR_2", "ERR_3"]
features_T278 = features_T274
x, y = preapre_data(json_path, test_name[1], features_TStroop, 1)
print(f" Размерность X: {x.shape}, размерность Y: {y.shape}")
```
